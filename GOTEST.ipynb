```python
from selenium import webdriver
import pandas as pd
import random
```

```python
chromedriver = './chromedriver_win32/chromedriver.exe'
test_url = "http://hi-psy.ru/elearning/test?id=15"
user_name = "628264"
pass_word = "L9ZHCI6L"
```

```python
driver = webdriver.Chrome(chromedriver)
driver.get(test_url)
username = driver.find_element_by_id("signin_username")
password = driver.find_element_by_id("signin_password")
username.send_keys(user_name)
password.send_keys(pass_word)
driver.find_element_by_id("signinForm").submit()
driver.get(test_url+"&go=1")
```

```python
quests = pd.DataFrame(columns = ["Id", "Text", "Option1", "Option2", "Option3","Option4", "Answer", "Sure"])
```

```python
#get questions
form = driver.find_element_by_id("questionForm")
while form:
    text = driver.find_element_by_class_name("testQuestionText").text
    qId = driver.find_element_by_name("questionId").get_attribute('value')

    rad0 = driver.find_element_by_id("radio-0")
    par = rad0.find_element_by_xpath('..')
    ans0 = par.text
    rad1 = driver.find_element_by_id("radio-1")
    par1 = rad1.find_element_by_xpath('..')
    ans1 = par1.text
    rad2 = driver.find_element_by_id("radio-2")
    par2 = rad2.find_element_by_xpath('..')
    ans2 = par2.text
    ans3 = ""
    try:
        rad3 = driver.find_element_by_id("radio-3")
        par3 = rad3.find_element_by_xpath('..')
        ans3 = par3.text
        ind = random.choice(range(4))
    except:
        ind = random.choice(range(3))
    driver.execute_script('arguments[0].removeAttribute("checked");', rad0)
    driver.execute_script('arguments[0].classList.remove("r_on");', par)
    rad = driver.find_element_by_id("radio-"+str(ind))
    driver.execute_script('arguments[0].setAttribute("checked","");', rad)
    par = rad.find_element_by_xpath('..')
    driver.execute_script('arguments[0].classList.add("r_on");', par)
    quests = quests.append({"Id": qId, "Text": text, "Option1": ans0, "Option2": ans1, "Option3": ans2,"Option4": ans3, "Answer": ind, "Sure": False}, ignore_index=True)
    form.submit()
    form = driver.find_element_by_id("questionForm")
```
